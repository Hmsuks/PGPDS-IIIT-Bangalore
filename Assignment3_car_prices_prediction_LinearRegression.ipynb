## Step 2. Data Cleaning

````python
cars_prices.info()
````

````python
# converting symboling to categorical
cars_prices['symboling'] = cars_prices['symboling'].astype('object')
cars_prices.info()
````

````python
cars_prices.head(10)
````

````python
#Filtering car brand from CarName
cars_prices['car_brand']=cars_prices['CarName'].apply(lambda x: x.split(" ")[0])
````

````python
cars_prices['car_brand'].head(30)
````

````python
# look at all values 
cars_prices['car_brand'].astype('category').value_counts()
````

````python
# correcting misspelled car_brand names

# volkswagen
cars_prices.loc[(cars_prices['car_brand'] == "vw") | 
         (cars_prices['car_brand'] == "vokswagen")
         , 'car_brand'] = 'volkswagen'

# porsche
cars_prices.loc[cars_prices['car_brand'] == "porcshce", 'car_brand'] = 'porsche'

# toyota
cars_prices.loc[cars_prices['car_brand'] == "toyouta", 'car_brand'] = 'toyota'

# nissan
cars_prices.loc[cars_prices['car_brand'] == "Nissan", 'car_brand'] = 'nissan'

# mazda
cars_prices.loc[cars_prices['car_brand'] == "maxda", 'car_brand'] = 'mazda'
````

````python
cars_prices['car_brand'].astype('category').value_counts()
````

````python
cars_prices.head(10)
````

````python
cars_prices.info()
````

````python
# dropping carname as we have created new column

# drop carname variable
cars_prices= cars_prices.drop('CarName', axis=1)
````

````python
#checking the categorical values in the dataset
cars_prices_categorical = cars_prices.select_dtypes(include=['object'])
````

````python
cars_prices_categorical.info()
````

````python
# convert categorical values into dummies
cars_prices_dummies = pd.get_dummies(cars_prices_categorical, drop_first=True)
cars_prices_dummies.head()
````

````python
# drop categorical variables 
cars_prices = cars_prices.drop(list(cars_prices_categorical.columns), axis=1)
````

````python
cars_prices.head()
````


## Step 4. Data Preparation


````python
cars_prices.head(10)
````

````python
# dropping carID variable 
cars_prices = cars_prices.drop(['car_ID'], axis=1)
````

````python
# concat dummy variables with X
cars_prices = pd.concat([cars_prices, cars_prices_dummies], axis=1)
````

````python
cars_prices.head()
````


### Rescaling the Features 
It is extremely important to rescale the variables so that they have a comparable scale. 
There are twocoon ways of rescaling 
1. Normalisation (min-max scaling) and 
2. standardisation (mean-o, sigma-1) 
Let's try normalisation

````python
#defining a normalisation function 
def normalize (x): 
    return ( (x-np.min(x))/ (max(x) - min(x)))
                                            
                                              
# applying normalize ( ) to all columns 
cars_prices = cars_prices.apply(normalize) 
````

````python
cars_prices.head()
````

````python
# verifying that after scaling the values are in between 0 and 1 
cars_prices.describe()
````

````python
# split into X and y

y = cars_prices.pop('price')
X = cars_prices
````

````python
# split into train and test
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    train_size=0.7,
                                                    test_size = 0.3, random_state=100)
````

## Step 5. Model Building and Evaluation

````python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
````

````python
# Building the first model with all the features

# instantiate
lm = LinearRegression()

# fit
lm.fit(X_train, y_train)
````

````python
y_pred = lm.predict(X_test)

# metrics
from sklearn.metrics import r2_score,mean_squared_error

print(r2_score(y_true=y_test, y_pred=y_pred))
````

we are getting 83.82% r-squared with all the variables. We will try to reduce the features to make the model simpler.

#### Model Building Using RFE

Initially we will build few models using recursive feature elimination to select features. We'll first start with n=15 arbitrary number of features, and then use the ```statsmodels``` library to build models using the shortlisted features.

````python
# RFE with 15 features
from sklearn.feature_selection import RFE

# RFE with 15 features
lm = LinearRegression()
rfe_15 = RFE(lm, 15)

# fit with 15 features
rfe_15.fit(X_train, y_train)

# making predictions using rfe model
y_pred = rfe_15.predict(X_test)

# r-squared
print(r2_score(y_test, y_pred))
````

RFE with 15 features is giving about 89.41% r-squared

````python
# RFE with 12 features
from sklearn.feature_selection import RFE

# RFE with 12 features
lm = LinearRegression()
rfe_12 = RFE(lm, 12)

# fit with 12 features
rfe_12.fit(X_train, y_train)

# making predictions using rfe model
y_pred = rfe_12.predict(X_test)

# r-squared
print(r2_score(y_test, y_pred))
````

RFE with 12 features is giving about 90.95% r-squared

````python
# RFE with 9 features
from sklearn.feature_selection import RFE

# RFE with 9 features
lm = LinearRegression()
rfe_9 = RFE(lm, 9)

# fit with 15 features
rfe_9.fit(X_train, y_train)

# making predictions using rfe model
y_pred = rfe_9.predict(X_test)

# r-squared
print(r2_score(y_test, y_pred))
````

RFE with 9 features is giving about 89.97% r-squared

````python
# RFE with 6 features
from sklearn.feature_selection import RFE

# RFE with 6 features
lm = LinearRegression()
rfe_6 = RFE(lm, 6)

# fit with 6 features
rfe_6.fit(X_train, y_train)

# predict
y_pred = rfe_6.predict(X_test)

# r-squared
print(r2_score(y_test, y_pred))
````

RFE with 6 features is giving about 88.96% r-squared

````python
# RFE with 4 features
from sklearn.feature_selection import RFE

# RFE with 4 features
lm = LinearRegression()
rfe_4 = RFE(lm, 4)

# fit with 6 features
rfe_4.fit(X_train, y_train)

# predict
y_pred = rfe_4.predict(X_test)

# r-squared
print(r2_score(y_test, y_pred))
````

RFE with 4 features is giving about 86.52% r-squared

There is sudden dip of 2.5% in the r-squared value when we reduced the features from 6 to 4 . 
Lets check the  adjusted r-squared, which penalises a model for having more features, and thus weighs both the goodness of fit and model complexity. We will use statsmodels library.

````python
# import statsmodels
import statsmodels.api as sm  

# subset the features selected by rfe_15
col_15 = X_train.columns[rfe_15.support_]

# subsetting training data for 15 selected columns
X_train_rfe_15 = X_train[col_15]

# add a constant to the model
X_train_rfe_15 = sm.add_constant(X_train_rfe_15)
lm_15 = sm.OLS(y_train, X_train_rfe_15).fit() 
print(lm_15.summary())
````

````python
# making predictions using rfe_15 sm model
X_test_rfe_15 = X_test[col_15]


# # Adding a constant variable 
X_test_rfe_15 = sm.add_constant(X_test_rfe_15)
X_test_rfe_15.info()


# # Making predictions
y_pred = lm_15.predict(X_test_rfe_15)
````

````python
# r-squared
print(r2_score(y_test, y_pred))
````

Thus, for the model with 15 variables, the r-squared on training and test data is about 93.9% and 89.4% respectively.There is significant difference of 4.5 % between train accuracy and test accuracy which may indicate that our model may 'overfit'. Lets reduce the number of features.

````python
#RFE 12

# import statsmodels
import statsmodels.api as sm  

# subset the features selected by rfe_12
col_12 = X_train.columns[rfe_12.support_]

# subsetting training data for 12 selected columns
X_train_rfe_12 = X_train[col_12]

# add a constant to the model
X_train_rfe_12 = sm.add_constant(X_train_rfe_12)
lm_12 = sm.OLS(y_train, X_train_rfe_12).fit() 
print(lm_12.summary())
````

````python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_test_rfe_6
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
````

### Final Model Evaluation

Evaluating the model in terms of Linear Regression assumptions
- The error terms are normally distributed with mean approximately 0

- Homoscedasticity, i.e. the 'spread' or 'variance' of the error term (y_true-y_pred) is constant

````python
y_pred = lm_5.predict(X_test_rfe_5)
````

````python
# Error terms
c = [i for i in range(len(y_pred))]
fig = plt.figure()
plt.plot(c,y_test-y_pred, color="blue", linewidth=2.5, linestyle="-")
fig.suptitle('Error Terms', fontsize=20)              # Plot heading 
plt.xlabel('Index', fontsize=18)                      # X-label
plt.ylabel('ytest-ypred', fontsize=16)                # Y-label
plt.show()
````

````python
# Plotting the error terms to understand the distribution.
fig = plt.figure()
sns.distplot((y_test-y_pred),bins=50)
fig.suptitle('Error Terms', fontsize=20)                  # Plot heading 
plt.xlabel('y_test-y_pred', fontsize=18)                  # X-label
plt.ylabel('Index', fontsize=16)                          # Y-label
plt.show()
````

````python
# mean
np.mean(y_test-y_pred)
````

Now it may look like that the mean is not 0, though compared to the scale of 'price', -0.011 is not such a big number

````python
# Plotting y_test and y_pred to understand the spread.
fig = plt.figure()
plt.scatter(y_test,y_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)              # Plot heading 
plt.xlabel('y_test', fontsize=18)                          # X-label
plt.ylabel('y_pred', fontsize=16)                          # Y-label
````

Spread or Variance is almost constant with some values spread out

## Preparing master_frame for sector analysis

```python
# Creating a new column primary sector and extract from category_list column
split_category = master_frame['category_list'].str.split('|',-1)
master_frame['primary_sector']=split_category.str.get(0)
master_frame.head()
```

## Sector Analysis 1: Determine main sector for each of the primary sector in master_frame

```python
# Import mapping.csv into mapping dataframe
mapping=pd.read_csv('mapping.csv', encoding='latin1')
mapping.head()
```

```python
# Identified an error in category_list column
mapping[(mapping['category_list']!='nan') & (mapping['category_list'].str.contains('0', regex = False))].head()
```

```python
# Handling error in  category_list column. '0' to be replaced with 'na'.
mapping['category_list']=mapping['category_list'].str.replace('0','na')
mapping[(mapping['category_list']!='nan') & (mapping['category_list'].str.contains('na', regex = False))].head()
```

```python
# We have a special case in category_list where one of the primary sector is enterptise 2.0. Need to handle this exception.
mapping['category_list']=mapping['category_list'].str.replace('Enterprise 2.na','Enterprise 2.0')
```

```python
# Unpivoting dataframe from short format to long format
mapping = (pd.melt(mapping, id_vars='category_list', var_name='map_sector'))
mapping.head(10)
```

```python
mapping['category_list']=mapping['category_list'].apply(lambda x: str(x).lower())
mapping.head()
```

```python
# Filter value column for 1's to determine main sector corresponding to each primary sector in categpory_list
mapping = mapping[mapping['value']==1]
mapping.head()
```

```python
# Creating a dictionary which handles key, value pairs of category_list & map_sector respectively
dict1 = mapping.set_index('category_list')['map_sector'].to_dict()
dict1
```

```python
# Converting main_sector column of master_frame to lower case
master_frame['primary_sector']=master_frame['primary_sector'].apply(lambda x: str(x).lower())
master_frame.head()
```

```python
# Creating column main_sector in master_frame dataframe and use the values in dict1 to fill the column using map function
master_frame['main_sector'] = master_frame['primary_sector'].map(dict1)
master_frame.head()
```

```python
# Identifying null values in the master_frame dataframe
master_frame.isnull().sum()
```

```python
# Removing rows having null values
master_frame = master_frame[~pd.isnull(master_frame.main_sector)]
master_frame.head()
```

## Sector Analysis 2: Creating data frames D1, D2 & D3.

### Creating D1

```python
# Creating data frame D1 where country = USA & $5 M<=funding<=& 15 M
D1 = master_frame[(master_frame.country_code == 'USA') & (master_frame.raised_amount_usd>=5000000) & 
                  (master_frame.raised_amount_usd<=15000000)]
D1.head()
```

```python
# Adding new column 'num_invest_sector' & sum_invest_sector
D1['num_invest_sector'] = D1.groupby('main_sector')['raised_amount_usd'].transform('count')
D1['sum_invest_sector'] = D1.groupby('main_sector')['raised_amount_usd'].transform('sum')
D1.head()
```

### Creating D2

```python
# Creating data frame D2 where country = GBR & $5 M<=funding<=& 15 M
D2 = master_frame[(master_frame.country_code == 'GBR') & (master_frame.raised_amount_usd>=5000000) & 
                  (master_frame.raised_amount_usd<=15000000)]
D2.head()
```

```python
# Adding new column 'num_invest_sector' & sum_invest_sector
D2['num_invest_sector'] = D2.groupby('main_sector')['raised_amount_usd'].transform('count')
D2['sum_invest_sector'] = D2.groupby('main_sector')['raised_amount_usd'].transform('sum')
D2.head()
```

### Creating D3

```python
# Creating data frame D3 where country = IND & $5 M<=funding<=& 15 M
D3 = master_frame[(master_frame.country_code == 'IND') & (master_frame.raised_amount_usd>=5000000) & 
                  (master_frame.raised_amount_usd<=15000000)]
D3.head()
```

```python
# Adding new column 'num_invest_sector' & sum_invest_sector
D3['num_invest_sector'] = D3.groupby('main_sector')['raised_amount_usd'].transform('count')
D3['sum_invest_sector'] = D3.groupby('main_sector')['raised_amount_usd'].transform('sum')
D3.head()
```
